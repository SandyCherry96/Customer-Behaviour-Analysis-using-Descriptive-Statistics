```python
# identify the outliers in income
plt.figure(figsize=(8,6))
sns.distplot(superstore_dataframe['Income'])
plt.title('Boxplot of Income')
plt.xlabel('Income')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

It shows that the DataFrame contains outliers. When dealing with outliers, it is recommended to use the median to fill missing values.

```python
# Fill NaN values in the 'Income' column with the median of that column
median_income = superstore_dataframe['Income'].median()
superstore_dataframe.fillna(median_income,inplace=True)
```

```python
# Calculate the first and third quartiles
Q1 = superstore_dataframe['Income'].quantile(0.25)
Q3 = superstore_dataframe['Income'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = superstore_dataframe[(superstore_dataframe['Income'] < lower_bound) | (superstore_dataframe['Income'] > upper_bound)]
print(len(outliers))
```

Any data points below the lower bound or above the upper bound are considered outliers. The factor of 1.5 is commonly used to determine outliers based on the IQR method.

```python
# Remove outliers
df_no_outliers = superstore_dataframe[(superstore_dataframe['Income'] >= lower_bound) & (superstore_dataframe['Income'] <= upper_bound)]
print(df_no_outliers)
```

# Recency: This variable could follow an Exponential distribution as it measures the time since the last purchase.

```python
import matplotlib.pyplot as plt

# Step 1: Visualize the Recency variable
recency_data = df_no_outliers['Recency'].dropna()

plt.hist(recency_data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Recency Histogram')
plt.xlabel('Recency (days)')
plt.ylabel('Density')
plt.show()
```

From the plot, it's clear that the Exponential distribution doesn't fit the Recency data well, as the actual data (in green) appears more uniformly spread. This visual reinforces that we should test other distributions, especially Uniform, which may be more suitable

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Step 1: Fit a Uniform distribution to the data
min_val = recency_data.min()
max_val = recency_data.max()

# Step 2: Generate the x values and Uniform PDF based on the fitted distribution
x = np.linspace(min_val, max_val, 1000)
pdf_uniform = stats.uniform.pdf(x, loc=min_val, scale=max_val - min_val)


# Step 3: Plot the Recency data histogram and the fitted Uniform distribution
plt.hist(recency_data, bins=30, density=True, alpha=0.6, color='g', label='Recency Data')
plt.plot(x, pdf_uniform, 'b-', lw=2, label='Uniform fit')

plt.title('Recency Data with Uniform Fit')
plt.xlabel('Recency (days)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

In the data, the column "Response" seems to be the binary variable representing whether a customer responded to the marketing campaign, where:

1 likely indicates a response.
0 likely indicates no response.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

prob_response = df_no_outliers['Response'].mean()

prob_success = stats.bernoulli.pmf(1, prob_response)
print(f"Probability of Response = 1 (success): {prob_success:.4f}")

expected_value = stats.bernoulli.mean(prob_response)
print(f"Expected value (mean) for Response: {expected_value:.4f}")

# Visualize the Bernoulli distribution
outcomes = [0, 1]
pmf_values = stats.bernoulli.pmf(outcomes, prob_response)

plt.bar(outcomes, pmf_values, align='center', alpha=0.5)
plt.xticks(outcomes)
plt.xlabel('Response')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution for Response')
plt.show()
```

```python
# NumDealsPurchases represents the count of purchases made over a certain time period and
# these purchases occur independently with a constant average rate,
# the Poisson distribution would be a suitable model for this variable.
```

```python
lambda_value = df_no_outliers['NumDealsPurchases'].mean()
print(f"Average number of deals purchased (lambda): {lambda_value:.2f}")

# Calculate the Poisson PMF for a range of possible deal purchases
k_values = np.arange(0, 10)  # Possible values (0 to 9 deals purchased)
pmf_values = stats.poisson.pmf(k_values, lambda_value)

# Create a bar plot for the Poisson distribution
plt.figure(figsize=(8, 6))
plt.bar(k_values, pmf_values, color='lightgreen', alpha=0.7, edgecolor='black')

# Customize plot aesthetics
plt.xticks(k_values)
plt.xlabel('Number of Deals Purchased', fontsize=14)
plt.ylabel('Probability', fontsize=14)
plt.title('Poisson Distribution of Number of Deals Purchased', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, max(pmf_values) * 1.1)  # Set y-axis limit

# Show the plot
plt.show()
```
